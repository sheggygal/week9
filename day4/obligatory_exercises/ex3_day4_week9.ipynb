```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_line, labs, ggtitle
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
data = pd.DataFrame({
    'x': x,
    'y1': y1,
    'y2': y2
})
```

```python
data_long = pd.melt(data, id_vars=['x'], value_vars=['y1', 'y2'],
                    var_name='Function', value_name='Value')
```

```python
plot = (ggplot(data_long, aes(x='x', y='Value', color='Function')) +
        geom_line() +
        labs(x='X-Value', y='Y-Value') +
        ggtitle('Comparison of Two Functions'))

print(plot)
```
