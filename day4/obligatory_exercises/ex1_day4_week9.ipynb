```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_point, labs, ggtitle
```

```python
import random
```

```python
np.random.seed(0)
data = pd.DataFrame({
    'x': np.random.rand(20),
    'y': np.random.rand(20)
})
```

```python
plot = (ggplot(data, aes(x='x', y='y')) +
        geom_point() +
        labs(x='X-Value', y='Y-Value') +
        ggtitle('Basic Scatter Plot'))

print(plot)
```
