```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_bar, labs, ggtitle
import random
```

```python
np.random.seed(0)
categories = ['A', 'B', 'C']
data = pd.DataFrame({
    'category': np.random.choice(categories, size=30)
})
```

```python
plot = (ggplot(data, aes(x='category')) +
        geom_bar() +
        labs(x='Category', y='Count') +
        ggtitle('Category Frequency'))

print(plot)
```
