```python
import seaborn as sns
import matplotlib.pyplot as plt
from plotnine import ggplot, aes, geom_point, labs, theme_bw
import plotly.express as px
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
print(iris.head())
```

```python
print(iris.shape)
```

```python
print(iris.isnull().sum())
```

```python
# Matplotlib: Histogram for petal lengths
plt.figure(figsize=(10, 6))
plt.hist(iris['petal_length'], bins=20, color='c', edgecolor='k')
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

The distribution is left-scewed, but we have another smaller peak to the right. Probably, that depends on a species of Iris.

```python
plt.figure(figsize=(10, 6))
species_list = iris['species'].unique()
colors = ['c', 'm', 'y']

for species, color in zip(species_list, colors):
    subset = iris[iris['species'] == species]
    plt.hist(subset['petal_length'], bins=20, alpha=0.5, color=color, edgecolor='k', label=species)

plt.title('Histogram of Petal Lengths by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.legend(title='Species')
plt.show()
```

This plot confirms, that the all the peaks are correlated sith the certain species.  

```python
# Seaborn: Pair plot to show pairwise relationships
sns.pairplot(iris, hue='species', palette='Set2')
plt.suptitle('Pairwise Relationships in the Iris Dataset', y=1.02)
plt.show()
```

Here we can see a clustered distribution of all the measures for all species.

```python
# Plotnine: Scatter plot of sepal length vs. sepal width
plot = (
    ggplot(iris, aes(x='sepal_length', y='sepal_width', color='species')) +
    geom_point() +
    labs(title='Sepal Length vs. Sepal Width', x='Sepal Length (cm)', y='Sepal Width (cm)') +
    theme_bw()
)
print(plot)
```

```python
# Plotly: Interactive 3D scatter plot
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', title='3D Scatter Plot of Iris Dataset',
                    labels={'sepal_length': 'Sepal Length (cm)', 'sepal_width': 'Sepal Width (cm)', 'petal_length': 'Petal Length (cm)'})
fig.show()
```

Here we can explore the relationship between three numerical variables - petal length, sepal length and a sepal width. Verginica has the biggest measurements, and setosa - the smallest.
